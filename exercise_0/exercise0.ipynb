## 3) Convert Between NumPy and Torch
- Create a NumPy array: `np.array([10, 20, 30])`
- Convert it to a torch tensor
- Convert that tensor back to NumPy
- Print all three objects and their Python types to compare

```python
# TODO
import numpy as np
import torch
np_array_orig = np.array([10,20,30])
torch_tensor = torch.from_numpy(np_array_orig)
np_array_new = torch_tensor.numpy()
print(f"Numpy orig: {np_array_orig} Type: {type(np_array_orig)}")
print(f"Torch tensor: {torch_tensor} Type: {type(torch_tensor)}")
print(f"Numpy new: {np_array_new} Type: {type(np_array_new)}")

```

## 4) Reshape
- Create a 1D tensor `t = torch.arange(12)`
- Reshape it to `(3, 4)`
- Reshape it back to `(12,)`
- Print the tensor values and all intermediate shapes

```python
# TODO
t = torch.arange(12)
t_D2 = t.reshape(3,4 )
t_D1 = t_D2.reshape(12)
print(f"t: {t}, Shape: {t.shape}")
print(f"t_D2: {t_D2}, Shape: {t_D2.shape}")
print(f"t_D1: {t_D1}, Shape: {t_D1.shape}")
```

## 5) Permute Dimensions
- Create `img = torch.rand(3, 4, 5)` and treat dimensions as `(C, H, W)`
- Permute dimensions to `(H, W, C)` with `permute(...)`
- Print original and permuted shapes
- Optional check: print one element index before/after to see only axes changed, not values

```python
# TODO
img_chw = torch.rand(3, 4, 5)
img_hwc = img_chw.permute(1, 2 ,0)
print(f"Img CHW: {img_chw}, Shape: {img_chw.shape}")
print(f"Img HWC: {img_hwc}, Shape: {img_hwc.shape}")
```

## 6) Squeeze and Unsqueeze
- Create `z` with shape `(1, 3, 1, 5)` (you can use `torch.randn(...)`)
- Use `squeeze` to remove dimensions of size 1
- Use `unsqueeze` to add one dimension back at `dim=0`
- Print the shape after each step to track how rank changes

```python
# TODO
z = torch.randn(1,3,1,5)
z_squeezed = z.squeeze()
z_unsqueezed = z_squeezed.unsqueeze(dim=0)

print(f"z: {z.shape}")
print(f"z_squeezed: {z_squeezed.shape}")
print(f"z_unsqueezed: {z_unsqueezed.shape}")
```
